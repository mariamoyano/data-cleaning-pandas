```python
import pandas as pd
import numpy as np
import re
```

```python
df_analysis = pd.read_csv("OUTPUT/shark-attacks.csv")
```

```python
df_analysis.head()
```

Hipotesis: ataques por paises y por decadas, ataques mortales y no mortales

# ATAQUES POR DÉCADAS:

Ataques no mortales por decadas:

```python
#Gráfico que representa los ataques no mortales por décadas
f=df_analysis.loc[df_analysis["Fatal (Y/N)"] == "N"].groupby("Decade").agg({"Fatal (Y/N)": "count"}).plot(kind='bar', stacked=True, figsize=(12,8),title = "ATAQUES NO MORTALES POR DÉCADAS")
```

```python
df_analysis.loc[df_analysis["Fatal (Y/N)"] == "N"].groupby("Decade").agg({"Fatal (Y/N)": "count"})
```

Ataques mortales por décadas:

```python
#Gráfico que representa los ataques no mortales por décadas
f0 = df_analysis.loc[df_analysis["Fatal (Y/N)"] == "Y"].groupby("Decade").agg({"Fatal (Y/N)": "count"}).plot(kind='bar', stacked=True, figsize=(12,8),title = "ATAQUES MORTALES POR DÉCADAS")
```

```python
df_analysis.loc[df_analysis["Fatal (Y/N)"] == "Y"].groupby("Decade").agg({"Fatal (Y/N)": "count"})
```
