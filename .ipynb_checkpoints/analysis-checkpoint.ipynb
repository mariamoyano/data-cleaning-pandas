```python
import pandas as pd
import numpy as np
import regex as re
```

```python
df_analysis = pd.read_csv("../shark-attacks.csv")
```

```python
#Group by decades
df_analysis.loc[df_analysis["Fatal (Y/N)"] == "N"].groupby("Decade").agg({"Fatal (Y/N)": "count"}).plot.bar()
```

```python
df_analysis.loc[df_analysis["Fatal (Y/N)"] == "Y"].groupby("Decade").agg({"Fatal (Y/N)": "count"}).plot.bar()
```

```python
df_analysis = df_analysis.sort_values(by=['Fatal (Y/N)'])
df_analysis.groupby("Country").agg({"Fatal (Y/N)": "count"}).plot.bar()
```

```python

df_analysis.groupby("Country").agg({"Fatal (Y/N)":"count", "Decade":"count"}).plot.bar()
```
