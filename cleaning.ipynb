```python
def fatal_clean(e):
   
    if e == ' N' or e == 'N':
        return 'N'
    elif e == 'y' or e == 'Y':
        return 'Y'
    else:
        return 'UNKNOWN'

df_clean["Fatal (Y/N)"] = df_clean["Fatal (Y/N)"].apply(fatal_clean)
```
